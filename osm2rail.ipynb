**Step 2: download OSM data in .osm format by given area name(s)**

```python
import osm2rail as orl
subarea_name='Grand Central Terminal'
download_dir='osmfile'
path_file=orl.download_osm_data_from_overpass(subarea_names=subarea_name,download_dir=download_dir,ret_download_path=True)
```

**Step 3: convert OSM data in .osm format to GMNS-rail network files**

```python
net=orl.get_network_from_OSMFile(osm_filename=path_file[0],strict_mode=True,POIs=True)
```

**Step 4: visualize rail network data set**

```python
orl.showNetwork(net)
```

**Step 5: output network to csv files**

```python
orl.saveNetwork(net)
```
